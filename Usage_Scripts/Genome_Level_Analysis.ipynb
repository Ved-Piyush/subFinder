```python
from pandas.errors import ParserError
```

```python
bad_paths = []
good_paths = []
ifdp_outputs_catch = []
jf_output_catch = []
cgc_data_catch = []

for clade in tqdm(clades):
    
    curr_clade_path = clade_path + "\\" + clade
    gcfs = os.listdir(curr_clade_path)
    
    for gcf in tqdm(gcfs): 
        
        curr_gcf_path = curr_clade_path + "\\" + gcf
        gcfs_content = os.listdir(curr_gcf_path)
        ifdp_outputs = pd.read_csv(curr_gcf_path + "\\" + '.output_counts_IFDP.tsv')
        ifdp_outputs.columns = ["Substrate", "Relative abundance"]
        ifdp_outputs["Clade"] = clade
        ifdp_outputs["GCF"] = gcf 
        ifdp_outputs_catch.append(ifdp_outputs)
        
        cgc_file_name = [item for item in gcfs_content if "concent" in item][0]
        jf_output = pd.read_csv(curr_gcf_path + "\\" + 'sub.prediction.out', sep = "\t")
        jf_output["Clade"] = clade
        jf_output["GCF"] = gcf 
        jf_output_catch.append(jf_output)
        
        try:
            path = curr_gcf_path + "\\" + cgc_file_name
            cgc_data = pd.read_csv(curr_gcf_path + "\\" + cgc_file_name, sep = " ", header = None)
            cgc_data["Clade"] = clade
            cgc_data["GCF"] = gcf
            good_paths.append(path)
            
        except ParserError:
            bad_paths.append(path)
            cgc_data = pd.DataFrame([[item[0], "".join(item[1:])]for item in pd.read_table(path, sep = "\t", header = None)[0].map(lambda x: x.split(" ")).values])
            cgc_data["Clade"] = clade
            cgc_data["GCF"] = gcf
            
        cgc_data_catch.append(cgc_data)

            
```

```python
cgc_data_catch_df = pd.concat(cgc_data_catch)
```

```python
cgc_data_catch_df = cgc_data_catch_df.drop([2], 1)
```

```python
jf_output_catch_df = pd.concat(jf_output_catch)
```
