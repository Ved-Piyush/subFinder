```python
weights = 1/(y_train.sum(0)/y_train.sum(0).sum())
```

```python
class_weights = {i: w for i, w in enumerate(weights)}
```

```python
def non_recurrent_attention_model(num_classes): 
    embedding_layer = tf.keras.layers.Embedding(len(weight_vectors),
                            weight_vectors.shape[1],
                            weights=[weight_vectors],
                            mask_zero = False,
                            trainable=False)
    
    vectorize_layer = tf.keras.layers.TextVectorization(
                     output_mode='int',
                     vocabulary=model_cbow.wv.index_to_key, 
                     standardize = None)
    input_layer = tf.keras.layers.Input(shape = (1,), dtype = tf.string)

    vectorize = vectorize_layer(input_layer)

    vectorize.shape

    emb_output = embedding_layer(vectorize)

    emb_output.shape

#     lstm_layer = tf.keras.layers.LSTM(100, return_sequences = True, dropout = 0.5)

#     lstm_output = lstm_layer(emb_output)

    x_a = tf.keras.layers.Dense(emb_output.get_shape()[-1]//6, kernel_initializer = 'glorot_uniform', activation="tanh", name="tanh_mlp")(emb_output) 
    
    x_a = tf.keras.layers.Dropout(0.6)(x_a)
    
    x_a = tf.keras.layers.Dense(1, kernel_initializer = 'glorot_uniform', activation='linear', name="word-level_context")(x_a)

    x_a = tf.keras.layers.Flatten()(x_a)

    att_out = tf.keras.layers.Activation('softmax')(x_a) 

    x_a2 = tf.keras.layers.RepeatVector(emb_output.get_shape()[-1])(att_out)

    x_a2 = tf.keras.layers.Permute([2,1])(x_a2)

    out = tf.keras.layers.Multiply()([emb_output,x_a2])
    
    out = tf.keras.layers.Lambda(lambda x : tf.math.reduce_sum(x, axis = 1), name='expectation_over_words')(out)
    
#     out = tf.keras.layers.Lambda(lambda x : K.sum(x, axis=1), name='expectation_over_words')(out)


    dropout_layer = tf.keras.layers.Dropout(0.6)(out)

    pred_head = tf.keras.layers.Dense(num_classes)

    pred_output = pred_head(dropout_layer)

    model = tf.keras.models.Model(input_layer, pred_output)
    
    model.compile(loss = tf.keras.losses.CategoricalCrossentropy(from_logits=True), 
                 optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-3), 
                 metrics= "accuracy")
    
    return model
```

```python
non_recurrent_attention_mod = non_recurrent_attention_model(y_train.shape[1])
```

```python
prob_threshold = 0.85
```

```python
safe_preds = uhgg_data[uhgg_data["predicted_prob"] >= prob_threshold]
```

```python
safe_preds["predicted_high_level_substr"].value_counts()
```

```python
imp_model = tf.keras.models.Model(non_recurrent_attention_mod.input, non_recurrent_attention_mod.layers[7].output)
```

```python
imp_model.summary()
```

```python
safe_preds = safe_preds.reset_index(drop = True)
```

```python
common_genes_all = []
common_gene_scores_all = []
yes_or_no = []
for i in tqdm(range(0, safe_preds.shape[0])): 
    
    gene_vec = safe_preds["sequence"][i]
    gene_vec = [gene_vec.replace("|", ",").replace(",", " ")]
    pred_prob = imp_model.predict(gene_vec, verbose = 0)[0]
    arg_sorts = np.argsort(-pred_prob)
    first_arg = arg_sorts[0]
    gene_vec_og = [gene for gene in gene_vec[0].split(" ")]
    gene_vec = [gene.split("_")[0] for gene in gene_vec[0].split(" ")]
    first_gene = gene_vec_og[first_arg]
    
    predicted_substrate = safe_preds["predicted_high_level_substr"][i]
    
    if predicted_substrate in markers_df["Substrate_high_level"].unique():
        markers_substrate = markers_df[markers_df["Substrate_high_level"] == predicted_substrate]
        all_markers = markers_substrate["Family"]
        any_common = list(set(all_markers.values[0]).intersection(gene_vec))
        here_idx = [i for i, gene in enumerate(gene_vec) if gene in any_common]
        common_gene = np.array(gene_vec)[here_idx]
        common_genes_all.append(common_gene.tolist())
        common_gene_scores = pred_prob[here_idx]
        common_gene_scores_all.append(common_gene_scores.tolist())
        yes_or_no.append("YES")
    
    else:
        catch_all = []
        for gene in all_markers.values[0]: 
            if gene in model_cbow.wv.index_to_key: 
                catch_all.append(model_cbow.wv.get_vector(gene))
        catch_all = np.array(catch_all).mean(0)
        first_gene_vec = model_cbow.wv.get_vector(first_gene)
        similarity_first_gene_all_markers = 1 - spatial.distance.cosine(catch_all, first_gene_vec)
        common_genes_all.append([first_gene])
        common_gene_scores_all.append([similarity_first_gene_all_markers])
        yes_or_no.append("NO")
```
